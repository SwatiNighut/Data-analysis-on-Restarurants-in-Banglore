```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
# some columns are not much useful so we will delete that columns.
```

```python
df.drop("url",axis="columns",inplace=True) # we are not analysing this that is why deleted
```

```python
df.drop("menu_item",axis="columns",inplace=True) # we are not analysing this that is why deleted
```

```python
df.drop('reviews_list',axis = 'columns',inplace= True) # we are not analysing this that is why deleted
```

```python
df.drop('address',axis = 'columns',inplace= True) # location column is there.
```

```python
df.drop('listed_in(city)',axis = 'columns',inplace= True) # location column is there.
```

```python
df.head()
```

```python
# we will rename some columns to it will be easy to understand.
```

```python
df.rename(columns={'name':'Resto_name','online_order':'Online_order','book_table':'Bookingtable','rate':'Rating','votes':'Votes','phone':'Ph_no','loctaion':'Location','rest_type':'Rest_type','dish_liked':'Dish_liked','cuisines':'Cuisines','approx_cost(for two people)':'Cost_2people(approx)','listed_in(type)':'Res_type'},inplace = True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Which are the top Restaurant in Banglore.

```python
df.Resto_name.value_counts().head(25).plot(kind= "bar")  # bar plot give clear about top 25 restaurant
plt.title("Top Restaurant")
plt.ylabel("Count")
plt.xlabel("Resto_name")
plt.show()
```

From Bar Graph we came to know that Cafe coffee day is the famous restaurent in banglore around 96 counts.
followed by Onesta and Just Bake.

# How the customer Rating is Distributed?

```python
cost_dist=df[['Rating','Cost_2people(approx)','Online_order']].dropna()
cost_dist['Rating']=cost_dist['Rating'].apply(lambda x: float(x.split('/')[0])if len(x)>3 else 0)
```

```python
sns.distplot(cost_dist['Rating'])
plt.show()
```

Most of restaurants have rating between 3-4.

# How Many Restaurant accept online orders?

```python
df['Online_order'].value_counts().plot.pie(autopct = "%1.1F%%", explode= (0.1,0))
plt.title('online order')
plt.show()
```

58.9% Restaurant accept online orders.

# How many Restaurant have facility of Table Booking.

```python
df['Bookingtable'].value_counts().plot.pie(autopct = "%1.1F%%", explode= (0,0.1))
plt.title('BookTable')
plt.show()
```

Only 12.5% Restaurant have facilty of Table booking.

# What is approx cost of 2 people for eating.

```python
cost_dist=df[['Rating','Cost_2people(approx)','Online_order']].dropna()
cost_dist['Rating']=cost_dist['Rating'].apply(lambda x: float(x.split('/')[0]) if len(x)>3 else 0)
cost_dist['Cost_2people(approx)']=cost_dist['Cost_2people(approx)'].apply(lambda x: int(x.replace(',','')))
```

```python
plt.figure(figsize=(6,6))
sns.distplot(cost_dist['Cost_2people(approx)'])


plt.show()
```

Approx cost of 2 people for eating is lies between 300-1000.

# Relation Between Approx cost of 2 people and Rating

```python
cost_dist=df[['Rating','Cost_2people(approx)','Online_order']].dropna()
cost_dist['Rating']=cost_dist['Rating'].apply(lambda x: float(x.split('/')[0]) if len(x)>3 else 0)
cost_dist['Cost_2people(approx)']=cost_dist['Cost_2people(approx)'].apply(lambda x: int(x.replace(',','')))
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="Rating",y='Cost_2people(approx)',hue='Online_order',data=cost_dist)
plt.show()
```

Approx cost for 2 people and rating relation.
Most of restaurant serve food less than 1000rs.
